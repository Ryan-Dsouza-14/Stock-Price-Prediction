# Procter & Gamble Stock Price Prediction

```python
#Importing necessary libraries
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
from scipy.stats import norm
%matplotlib inline
```

```python
import yfinance as yf   #Extracting Stock data from Yahoo Finance

PG = yf.download('PG', start='2007-01-01')
PG
```

```python
PG = yf.download('PG', start='2007-01-01', group_by='ticker')['Adj Close']
PG
```

```python
log_returns = np.log(1+ PG.pct_change())
```

```python
log_returns
```

```python
log_returns.tail()
```

```python
PG.plot(figsize=(10,6))  #ploting stock data from 2007
```

```python
log_returns.plot(figsize=(10,6))   #ploting log returns of the stock
```

```python
u = log_returns.mean()
u
```

```python
var = log_returns.var()
```

```python
var
```

```python
drift = u - (0.5 * var)
drift
```

```python
stdev = log_returns.std()
stdev
```

```python
t_intervals = 1000
iterations = 10
```

```python
daily_returns = np.exp(np.array(drift) + np.array(stdev) * norm.ppf(np.random.rand(t_intervals,iterations)))
daily_returns
```

```python
S0 = PG.iloc[-1]
print(S0)
```

```python
price_list = np.zeros_like(daily_returns)
price_list
```

```python
price_list[0] = S0
price_list
```

```python
for t in range(1, t_intervals):
    price_list[t]= price_list[t-1] * daily_returns[t]
```
